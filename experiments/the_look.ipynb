```python
from preql import Dialects
from trilogy_public_models import models
env = models["bigquery.thelook_ecommerce"]

executor = Dialects.BIGQUERY.default_executor(environment=env)
```

```python

QA_3 = """
property order_items.id.revenue <- order_items.sale_price*orders.num_of_item;
# shipped, complete, processing
key order_items.bookable_revenue <- filter order_items.revenue where order_items.status NOT IN ('Cancelled','Returned' );
key orders.bookable <- filter orders.id where orders.status NOT IN ('Cancelled','Returned' );
SELECT 
    date_trunc(orders.created_at, month) -> orders.creation_month,
    sum(order_items.bookable_revenue) -> revenue,
    count(orders.bookable) -> orders,
ORDER BY orders.creation_month desc;
"""

results = executor.execute_text(QA_3)
```

```python
for r in results[0]:
    print(r)
```

```python

QA_4 = """
property products.id.total_sales <- sum(order_items.sale_price) by products.id;
property products.id.total_profit <- sum(order_items.sale_price - products.cost) by products.id;
property products.id.sales_rank <- rank products.id by products.total_profit desc;

SELECT 
    products.name,
    products.total_sales,
    products.total_profit,
    products.sales_rank
ORDER BY products.sales_rank
asc
limit 100;
"""

results = executor.execute_text(QA_4)
```

```python
for r in results[0]:
    print(r)
```
